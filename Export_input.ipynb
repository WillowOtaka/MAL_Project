```python
import json
import numpy as np
import pandas as pd
```

```python
input_no_info = pd.read_csv('./MAL_data/input_no_info.csv')
input_no_info
```

```python
animes_data = pd.read_csv('./MAL_data/animes.csv')
animes_data = animes_data.drop_duplicates()
animes_data
```

```python
anime_columns = [f'a{n}' for n in range(10)] + ['ar']
for column in anime_columns:
    letter = str(column[-1])
    input_no_info = input_no_info.rename(columns={column: 'uid'})
    input_no_info = input_no_info.merge(animes_data[['uid', 'genre', 'episodes', 'score']], on='uid')
    input_no_info = input_no_info.rename(columns={'uid': column, 'genre': f'genre{letter}', 'episodes': f'episodes{letter}', 'score': f'score{letter}'})
input_no_info
```

```python
anime_columns = ['a0', 'genre0', 'episodes0', 'score0',
                 'a1', 'genre1', 'episodes1', 'score1',
                 'a2', 'genre2', 'episodes2', 'score2',
                 'a3', 'genre3', 'episodes3', 'score3',
                 'a4', 'genre4', 'episodes4', 'score4',
                 'a5', 'genre5', 'episodes5', 'score5',
                 'a6', 'genre6', 'episodes6', 'score6',
                 'a7', 'genre7', 'episodes7', 'score7',
                 'a8', 'genre8', 'episodes8', 'score8',
                 'a9', 'genre9', 'episodes9', 'score9',
                 'ar', 'genrer', 'episodesr', 'scorer',
                 'like']
input_df = pd.DataFrame({column: input_no_info[column] for column in anime_columns})
input_df
```

```python
input_df.to_csv('./MAL_data/input.csv')
```
